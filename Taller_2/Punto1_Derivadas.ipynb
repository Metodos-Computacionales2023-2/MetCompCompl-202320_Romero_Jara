```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
r = 0.25
R = 1750
Bo = 0.05
f = 7
w = 3.5
T = (2*np.pi)/w
t = np.linspace(0,2*T,1000)
```

```python
def Derivada(f,x,h=1e-8):
    return (f(x+h)-f(x-h))/(2*h)

def Flujo(t):
    return np.pi*(r**2)*Bo*np.cos(w*t)*np.cos(2*np.pi*f*t)
def Corriente_Inducida(t):
    return (-1)*(1/R)*Derivada(Flujo,t)
```

```python
I = Corriente_Inducida(t)
De_I = Derivada(Corriente_Inducida,t)
```

```python
fig,ax = plt.subplots()
ax.plot(t,I)
ax.axhline(y=0, color='red')
ax.set_ylabel('Corriente inducida sobre el bucle')
ax.set_xlabel('Tiempo')
ax.set_title('Corriente inducida')
```
